```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv("./data/california_cities.csv")
cities.head()
```

```python
cities.tail()
```

```python
# Extract Latitude and Longtitude
```

```python
latd = cities["latd"]
longd = cities["longd"]
pop = cities["population_total"]
area = cities["area_total_km2"]
```

```python
print(latd[:10])
```

```python
plt.style.use("seaborn-v0_8")
plt.figure(figsize=(8,7))
plt.scatter(longd, latd, c = np.log10(pop), cmap="viridis", s = area, linewidths=0, alpha=0.5)
plt.axis("equal")
plt.xlabel("longtitude")
plt.ylabel("lattitude")
plt.colorbar(label="log$_{10}$(pop)")
plt.clim(3,7)

# Area
area_range = [50, 100, 300, 500]
for i in area_range:
    plt.scatter([], [], s = i, c='k', alpha=0.4 , label=str(i)+"km$^2$")

plt.legend(labelspacing=1, title="Areas")


plt.title("California Cities: Populations and Areas");
```
